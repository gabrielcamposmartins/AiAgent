```python
plt.figure(figsize=(8, 8), facecolor='#1d2430', edgecolor='white')
plt.axes(facecolor='#1d2430')
plt.plot([1, 2, 3, 4, 6], [1, 4, 2, 3, 5], color='green')
plt.plot([1, 2, 3, 4, 6], [1, 3, 2, 4, 5], color='red')
plt.title('My first plot')
plt.show() 
```

```python
plt.figure(figsize=(8, 8), facecolor='#1d2430', edgecolor='white')
plt.rcParams["figure.figsize"] = [7.00, 3.50]
plt.rcParams["figure.autolayout"] = True
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4)
for ax, color in zip([ax1, ax2, ax3, ax4], ['green', 'red', 'yellow', 'blue']):
   plt.setp(ax.spines.values(), color=color)
   ax.plot([8, 3], c=color)
   plt.setp([ax.get_xticklines(), ax.get_yticklines()], color=color)
plt.show()
```

```python
from sklearn.datasets import make_classification, make_blobs
```

```python
plt.title("Three blobs", fontsize="small")
X1, Y1 = make_blobs(n_features=2, centers=3)
plt.scatter(X1[:, 0], X1[:, 1], marker="o", c=Y1, s=25, edgecolor="k")
```

```python
# plt.subplot(322)
plt.title("Two informative features, one cluster per class", fontsize="small")
X1, Y1 = make_classification(
    n_features=2, n_redundant=0, n_informative=2, n_clusters_per_class=1
)
plt.scatter(X1[:, 0], X1[:, 1], marker="o", c=Y1, s=25, edgecolor="k")
```
